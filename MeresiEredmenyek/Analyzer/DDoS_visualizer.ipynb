#NoContent

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file = 'log_na_nc.csv'
df = pd.read_csv(file)

df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S')
reference_time = df['Time'].min()
df['ElapsedSeconds'] = (df['Time'] - reference_time).dt.total_seconds()

fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot Response Time on the first y-axis
ax1.plot(df['ElapsedSeconds'], df['ResponseTimeInms'], label='Average response time (ms)', color='blue', linewidth=2, alpha=0.8)
ax1.set_ylabel('Response time (ms)', color='black', fontsize=12)
ax1.tick_params(axis='y', labelcolor='black')
ax1.set_ylim(bottom=0)

ax1.grid(visible=True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)

# Plot Time
ax1.set_xlabel('Time (seconds)', fontsize=12)
max_seconds = df['ElapsedSeconds'].max()
xticks = np.arange(0, max_seconds + 20, 40)
plt.xticks(xticks)

# Add a legend
lines1, labels1 = ax1.get_legend_handles_labels()
fig.legend(lines1, labels1, loc="upper center", bbox_to_anchor=(0.5, 0.95), ncol=4, fontsize=10, frameon=False)

# Add a title and adjust the layout
plt.title('ESP32 server response times without stressing with NoContent requests ', fontsize=14)
fig.tight_layout()

# Rotate x-axis labels for better readability
plt.xticks(rotation=0, fontsize=10)
plt.yticks(fontsize=10)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file1 = 'log_a_pi.csv'
df1 = pd.read_csv(file1)

file2 = 'log_pi_attacker.csv'
df2 = pd.read_csv(file2)

df1['Time'] = pd.to_datetime(df1['Time'], format='%H:%M:%S')
df2['Time'] = pd.to_datetime(df2['Time'], format='%H:%M:%S')
merged_df = pd.merge(df1, df2, on='Time', how='outer')

reference_time = merged_df['Time'].min()
merged_df['ElapsedSeconds'] = (merged_df['Time'] - reference_time).dt.total_seconds()

# Interpolate between meassure points
merged_df['ResponseTimeInms'] = merged_df['ResponseTimeInms'].interpolate()
merged_df['SentRequestsPerSecond'] = merged_df['SentRequestsPerSecond'].interpolate()
merged_df['SusceededResponses'] = merged_df['SusceededResponses'].interpolate()
merged_df['TimedoutResponses'] = merged_df['TimedoutResponses'].interpolate()

fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot Response Time on the first y-axis
ax1.plot(merged_df['ElapsedSeconds'], merged_df['ResponseTimeInms'], label='Average response time (ms)', color='blue', linewidth=2, alpha=0.8)
ax1.set_ylabel('Response time (ms)', color='black', fontsize=12)
ax1.tick_params(axis='y', labelcolor='black')
ax1.set_ylim(bottom=0)

ax1.grid(visible=True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
ax2 = ax1.twinx()

# Plot Sent Requests Per Second
ax2.plot(merged_df['ElapsedSeconds'], merged_df['SentRequestsPerSecond'], label='Sent requests/sec', color='magenta', linewidth=2, alpha=1)
ax2.plot(merged_df['ElapsedSeconds'], merged_df['SusceededResponses'], label='Succeeded responses count', color='orange', linewidth=1.5, alpha=0.8)
ax2.plot(merged_df['ElapsedSeconds'], merged_df['TimedoutResponses'], label='Timed out responses count', color='red', linewidth=1.5, alpha=0.8)
ax2.set_ylabel('Requests count', color='black', fontsize=12)
ax2.tick_params(axis='y', labelcolor='black')
ax2.set_ylim(bottom=0)

# Plot Time
ax1.set_xlabel('Time (seconds)', fontsize=12)
max_seconds = merged_df['ElapsedSeconds'].max()
xticks = np.arange(0, max_seconds + 20, 40)
plt.xticks(xticks)

# Add a legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
fig.legend(lines1 + lines2, labels1 + labels2, loc="upper center", bbox_to_anchor=(0.5, 0.95), ncol=4, fontsize=10, frameon=False)

# Add a title and adjust the layout
plt.title('ESP32 server response times during stressing with Pi requests ', fontsize=14)
fig.tight_layout()

# Rotate x-axis labels for better readability
plt.xticks(rotation=0, fontsize=10)
plt.yticks(fontsize=10)

# Display the plot
plt.show()
```
